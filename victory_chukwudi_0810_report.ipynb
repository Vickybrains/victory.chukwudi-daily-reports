NumPy stands for Numerical Python and provides support for arrays, matrices, and a wide range of mathematical functions to operate on these data structures. This report summarizes the key features and functionalities of NumPy that I found particularly useful.
Key Features of NumPy

    Array Creation:
        NumPy allows for the creation of multi-dimensional arrays (ndarrays). I learned how to create arrays from lists, tuples, or using built-in functions like np.zeros(), np.ones(), and np.arange().
        Example: np.array([1, 2, 3]) creates a 1D array, while np.array([[1, 2], [3, 4]]) creates a 2D array.

    Array Operations:
        Element-wise operations can be performed easily on NumPy arrays. I learned how to add, subtract, multiply, and divide arrays using operators without the need for explicit loops.
        Example: Adding two arrays a + b directly adds corresponding elements.

    Array Manipulation:
        NumPy provides various functions for reshaping, stacking, and splitting arrays. I practiced reshaping arrays using reshape(), stacking them vertically or horizontally with vstack() and hstack(), and splitting them into sub-arrays using split().
        Example: np.vstack((array_a, array_b)) stacks arrays vertically.

    Mathematical Functions:
        I learned how to compute various mathematical operations such as sum, mean, median, standard deviation, and more using NumPy functions like np.sum(), np.mean(), np.median(), and np.std().
        Example: np.mean(array) computes the average of all elements in the array.

    Linear Algebra:
        The course introduced me to basic linear algebra operations using NumPy, including matrix multiplication with np.dot() and finding determinants and inverses using functions like np.linalg.det() and np.linalg.inv().
        Example: Multiplying two matrices C = np.dot(A, B).

    Random Number Generation:
        NumPy includes a module for generating random numbers, which is useful for simulations and testing. Functions like np.random.rand() and np.random.randint() allow for the creation of random arrays.
        Example: np.random.rand(3, 3) generates a 3x3 matrix of random floats.

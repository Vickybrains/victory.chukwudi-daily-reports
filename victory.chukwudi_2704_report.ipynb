2. Operations on Lists
The lecture detailed several fundamental operations that can be performed on lists:

Accessing Elements: Lists are zero-indexed, meaning the first element is accessed with index 0.

````python

print(my_list[0])  # Output: 1
Slicing: Lists can be sliced to obtain a sublist.

sublist = my_list[1:4]  # Output: [2, 3, 'hello']
Appending Elements: The append() method allows adding elements to the end of a list.

my_list.append(6)
Inserting Elements: The insert() method can add elements at a specific index.

my_list.insert(2, 'new')
Removing Elements: The remove() method deletes the first occurrence of a specified value.

my_list.remove('hello')
Iterating Over Lists: Lists can be iterated using loops.

for item in my_list:
    print(item)
````

One Argument: Generates numbers from 0 to the specified number (exclusive).

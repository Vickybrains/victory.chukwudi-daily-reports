````python
### Exercise 1:
Write a Python function `add_numbers(a, b)` that takes two numbers and returns their sum.
def add_numbers(a,b):
return a + b
result = add_numbers(7,8)
print(f"The sum of the numbers is:{result}")
```python
# Exercise 1
```

---

### Exercise 2:
Create a function `greet(name)` that takes a name and prints "Hello, <name>!".

```python
# Exercise 2
```
def greet(name):
    print(f"Hello,{name}!")
user_name = input("Enter your name: ")
greet(user_name)
---

### Exercise 3:
Write a Python function `is_even(n)` that checks whether a number is even.

```python
# Exercise 3
```
def is_even(n):
    return n % 2 == 0
number = int(input("Enter a number: "))
if is_even(number):
    print(f"{number} is even.")
else:
    print(f"{number} is odd.")
---

### Exercise 4:
Write a function `max_of_three(a, b, c)` that returns the largest of three numbers.

```python
# Exercise 4
```
def find_max_of_three(a, b, c):
    max_of_two = find_max(a, b)
    return find_max(max_of_two, c)
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

result = find_max_of_three(num1, num2, num3)
print(f"The maximum value among the three numbers is: {result}")


---

### Exercise 5:
Create a Python module `math_utils.py` and define a function `square(x)` that returns the square of a number.

```python
# Exercise 5
```
def square(x):
    """Return the square of a number."""
    return x * x
from math_utils import square

# Example usage
number = float(input("Enter a number: "))
result = square(number)
print(f"The square of {number} is {result}.")

---

### Exercise 6:
Write a function `factorial(n)` that returns the factorial of a number using recursion.

```python
# Exercise 6
```
def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
try:
    number = int(input("Enter a non-negative integer: "))
    result = factorial(number)
    print(f"The factorial of {number} is {result}.")
except ValueError as e:
    print(e)

---

### Exercise 7:
Write a function `is_prime(n)` that checks if a number is prime.

```python
# Exercise 7
```
def is_prime(n):
    if n <= 1:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check divisibility up to the square root of n
        if n % i == 0:
            return False  # If n is divisible by any number other than 1 and itself, it's not prime
    return True  # If no divisors were found, n is prime
number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")

---

### Exercise 8:
Write a function `sum_of_list(lst)` that takes a list of numbers and returns the sum of all the elements.

```python
# Exercise 8
```
def sum_of_list(lst):
    """Return the sum of all elements in the list."""
    total = 0
    for number in lst:
        total += number
    return total
numbers = [1, 2, 3, 4, 5]  
result = sum_of_list(numbers)
print(f"The sum of the list is: {result}")

---

### Exercise 9:
Write a function `fibonacci(n)` that returns the `n`th Fibonacci number using recursion.

```python
# Exercise 9
```
def fibonacci(n):
    """Return the n-th Fibonacci number."""
    if n < 0:
        raise ValueError("Input should be a non-negative integer.")
    elif n == 0:
        return 0 
    elif n == 1:
        return 1 
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
try:
    number = int(input("Enter a non-negative integer: "))
    result = fibonacci(number)
    print(f"The {number}th Fibonacci number is: {result}")
except ValueError as e:
    print(e)

---

### Exercise 10:
Create a function `reverse_string(s)` that takes a string and returns the reverse of the string.

```python
# Exercise 10
```
def reverse_string(s):
    """Return the reverse of the given string."""
    return s[::-1]
input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)
print(f"The reversed string is: {reversed_string}")

---

### Exercise 11:
Write a function `count_vowels(s)` that counts the number of vowels in a string.

```python
# Exercise 11
```
def count_vowels(s):
    """Count the number of vowels in the given string."""
    vowels = 'aeiouAEIOU'  
    count = 0
    
    for char in s:
        if char in vowels:
            count += 1
            
    return count
input_string = input("Enter a string: ")
vowel_count = count_vowels(input_string)
print(f"The number of vowels in the string is: {vowel_count}")

---

### Exercise 12:
Write a function `is_palindrome(s)` that checks whether a string is a palindrome.

```python
# Exercise 12
```
def is_palindrome(s):
    """Check if the given string is a palindrome."""
    normalized_str = s.replace(" ", "").lower()
    return normalized_str == normalized_str[::-1]
input_string = input("Enter a string: ")
if is_palindrome(input_string):
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome.')

---

### Exercise 13:
Write a function `fahrenheit_to_celsius(f)` that converts Fahrenheit to Celsius.

```python
# Exercise 13
```
def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9
fahrenheit_temp = float(input("Enter temperature in Fahrenheit: "))
celsius_temp = fahrenheit_to_celsius(fahrenheit_temp)
print(f"{fahrenheit_temp}°F is equal to {celsius_temp:.2f}°C.")

---

### Exercise 14:
Write a Python function `is_leap_year(year)` that checks if a year is a leap year.

```python
# Exercise 14
```
def is_leap_year(year):
    """Check if the given year is a leap year."""
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
try:
    year_input = int(input("Enter a year: "))
    if is_leap_year(year_input):
        print(f"{year_input} is a leap year.")
    else:
        print(f"{year_input} is not a leap year.")
except ValueError:
    print("Please enter a valid integer year.")

---

### Exercise 15:
Write a Python function `find_max_in_list(lst)` that returns the maximum value from a list.

```python
# Exercise 15
```
def find_max_in_list(lst):
    """Return the maximum value from the list."""
    if not lst:  
        raise ValueError("The list is empty.")
    
    max_value = lst[0]  
    for number in lst:
        if number > max_value:
            max_value = number  
            
    return max_value
try:
    numbers = [int(x) for x in input("Enter numbers separated by spaces: ").split()]
    max_value = find_max_in_list(numbers)
    print(f"The maximum value in the list is: {max_value}")
except ValueError as e:
    print(e)

---

### Exercise 16:
Write a Python function `remove_duplicates(lst)` that removes duplicates from a list.

```python
# Exercise 16
```
def remove_duplicates(lst):
    """Return a list with duplicates removed."""
    return list(set(lst))
input_list = input("Enter elements of the list separated by spaces: ").split()
unique_list = remove_duplicates(input_list)
print(f"The list after removing duplicates: {unique_list}")

---

### Exercise 17:
Write a Python function `sum_of_squares(n)` that returns the sum of squares of the first `n` natural numbers.

```python
# Exercise 17
```
def sum_of_squares(n):
    """Return the sum of squares of the first n natural numbers."""
    if n < 1:
        raise ValueError("Input must be a positive integer.")
    return sum(i**2 for i in range(1, n + 1))
try:
    number = int(input("Enter a positive integer: "))
    result = sum_of_squares(number)
    print(f"The sum of squares of the first {number} natural numbers is: {result}")
except ValueError as e:
    print(e)

---

### Exercise 18:
Write a Python function `check_password_strength(password)` that returns `True` if the password meets the length and complexity requirements.

```python
# Exercise 18
```
def check_password_strength(password):
    """Check if the password meets length and complexity requirements."""
    # Define password requirements
    min_length = 8
    has_upper = any(c.isupper() for c in password)  
    has_lower = any(c.islower() for c in password)  
    has_digit = any(c.isdigit() for c in password)  
    has_special = any(c in "!vic@aims" for c in password)  
    if (len(password) >= min_length and
        has_upper and
        has_lower and
        has_digit and
        has_special):
        return True
    return False
user_password = input("Enter your password: ")
if check_password_strength(user_password):
    print("Password is strong.")
else:
    print("Password is weak. It must be at least 8 characters long and include uppercase, lowercase, digits, and special characters.")

---

### Exercise 19:
Write a function `merge_lists(list1, list2)` that merges two lists into one.

```python
# Exercise 19
```
def merge_lists(list1, list2):
    return list1 + list2

# Example usage
list1 = input("Enter the first list of elements separated by spaces: ").split()
list2 = input("Enter the second list of elements separated by spaces: ").split()

merged_list = merge_lists(list1, list2)
print(f"The merged list is: {merged_list}")

---

### Exercise 20:
Write a function `find_factors(n)` that returns a list of all factors of a number `n`.

```python
# Exercise 20
```
def find_factors(n):
    """Return a list of all factors of the number n."""
    if n <= 0:
        raise ValueError("Input must be a positive integer.")
    
    factors = []
    for i in range(1, n + 1):
        if n % i == 0:
            factors.append(i)
    
    return factors
try:
    number = int(input("Enter a positive integer: "))
    result = find_factors(number)
    print(f"The factors of {number} are: {result}")
except ValueError as e:
    print(e)

---

### Exercise 21:
Write a Python module `string_utils.py` that defines a function `capitalize_first_letter(s)` which capitalizes the first letter of each word in a string.

```python
# Exercise 21

def capitalize_first_letter(s):
    return ' '.join(word.capitalize() for word in s.split())
if __name__ == "__main__":
    input_string = input("Enter a string: ")
    result = capitalize_first_letter(input_string)
    print(f"The capitalized string is: {result}")

---

### Exercise 22:
Write a Python function `sum_even_numbers(lst)` that returns the sum of all even numbers in a list.

```python
# Exercise 22
```
def sum_even_numbers(lst):
    """Return the sum of all even numbers in the list."""
    return sum(num for num in lst if num % 2 == 0)
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [int(x) for x in input_list]

result = sum_even_numbers(numbers)
print(f"The sum of even numbers in the list is: {result}")

---

### Exercise 23:
Write a Python function `generate_random_number()` that returns a random number between 1 and 100.

```python
# Exercise 23
```import random

def generate_random_number():
    """Return a random number between 1 and 100."""
    return random.randint(1, 100)
random_number = generate_random_number()
print(f"The generated random number is: {random_number}")


---

### Exercise 24:
Write a Python function `get_even_numbers(lst)` that returns a list of all even numbers from a given list.

```python
# Exercise 24
```
def get_even_numbers(lst):
    """Return a list of all even numbers from the given list."""
    return [num for num in lst if num % 2 == 0]
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [int(x) for x in input_list]
even_numbers = get_even_numbers(numbers)
print(f"The even numbers in the list are: {even_numbers}")

---

### Exercise 25:
Write a Python function `count_occurrences(lst, item)` that counts how many times an item appears in a list.

```python
# Exercise 25
```
def count_occurrences(lst, item):
    """Count how many times an item appears in the list."""
    return lst.count(item)
input_list = input("Enter elements of the list separated by spaces: ").split()
item_to_count = input("Enter the item to count: ")
occurrences = count_occurrences(input_list, item_to_count)
print(f"The item '{item_to_count}' appears {occurrences} times in the list.")

---

### Exercise 26:
Write a Python function `gcd(a, b)` that calculates the greatest common divisor of two numbers using recursion.

```python
# Exercise 26
```
def gcd(a, b):
    """Calculate the greatest common divisor of two numbers using recursion."""
    if b == 0:
        return a
    return gcd(b, a % b)
try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = gcd(num1, num2)
    print(f"The GCD of {num1} and {num2} is: {result}")
except ValueError:
    print("Please enter valid integers.")

---

### Exercise 27:
Create a Python module `date_utils.py` and define a function `days_in_month(month, year)` that returns the number of days in a given month.

```python
# Exercise 27
```
def multiply_elements(lst, factor):
    """Multiply each element of the list by the given factor."""
    return [element * factor for element in lst]
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [float(x) for x in input_list]

factor = float(input("Enter the factor to multiply by: "))
result = multiply_elements(numbers, factor)

print(f"The list after multiplying by {factor} is: {result}")

---

### Exercise 28:
Write a Python function `convert_to_binary(n)` that converts an integer to its binary equivalent.

```python
# Exercise 28
```
# date_utils.py

def convert_to_binary(n):
    """Convert an integer to its binary equivalent."""
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")
    
    if n == 0:
        return "0"
    
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    
    return binary
try:
    number = int(input("Enter a non-negative integer: "))
    binary_representation = convert_to_binary(number)
    print(f"The binary equivalent of {number} is: {binary_representation}")
except ValueError as e:
    print(e)


---

### Exercise 29:
Write a Python function `multiply_elements(lst, factor)` that multiplies each element of a list by a given factor.

```python
# Exercise 29
```
def multiply_elements(lst, factor):
    """Multiply each element of the list by the given factor."""
    return [element * factor for element in lst]
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [float(x) for x in input_list]

factor = float(input("Enter the factor to multiply by: "))
result = multiply_elements(numbers, factor)

print(f"The list after multiplying by {factor} is: {result}")

---

### Exercise 30:
Write a Python function `replace_spaces(s)` that replaces spaces in a string with underscores.

```python
# Exercise 30
```
def replace_spaces(s):
    """Replace spaces in the string with underscores."""
    return s.replace(" ", "_")
input_string = input("Enter a string: ")
result = replace_spaces(input_string)
print(f"The string with spaces replaced by underscores is: {result}")

---

### Exercise 31:
Write a Python function `is_substring(s1, s2)` that checks whether `s1` is a substring of `s2`.

```python
# Exercise 31
```
def is_substring(s1, s2):
    """Check if s1 is a substring of s2."""
    return s1 in s2

# Example usage
string1 = input("Enter the first string (s1): ")
string2 = input("Enter the second string (s2): ")

if is_substring(string1, string2):
    print(f"'{string1}' is a substring of '{string2}'.")
else:
    print(f"'{string1}' is not a substring of '{string2}'.")
---

### Exercise 32:
Write a Python function `sum_of_digits(n)` that returns the sum of the digits of a given number.

```python
# Exercise 32
```
def sum_of_digits(n):
    return sum(int(digit) for digit in str(abs(n)))
try:
    number = int(input("Enter an integer: "))
    result = sum_of_digits(number)
    print(f"The sum of the digits of {number} is: {result}")
except ValueError:
    print("Please enter a valid integer.")

---

### Exercise 33:
Write a Python function `caesar_cipher(s, shift)` that implements the Caesar cipher encryption.

```python
# Exercise 33
```
def caesar_cipher(s, shift):
    encrypted = ""
    
    for char in s:
        if char.isalpha():
            offset = ord('A') if char.isupper() else ord('a')
            encrypted += chr((ord(char) - offset + shift) % 26 + offset)
        else:
            encrypted += char        
    return encrypted
input_string = input("Enter a string to encrypt: ")
shift_value = int(input("Enter the shift value: "))
result = caesar_cipher(input_string, shift_value)
print(f"The encrypted string is: {result}")

---

### Exercise 34:
Write a Python function `find_min_in_list(lst)` that returns the minimum value from a list.

```python
# Exercise 34
```
def find_min_in_list(lst):
    if not lst:  
        return None 

    minimum = lst[0]  
    for element in lst:
        if element < minimum:
            minimum = element  
            
    return minimum
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [float(x) for x in input_list]

result = find_min_in_list(numbers)
if result is not None:
    print(f"The minimum value in the list is: {result}")
else:
    print("The list is empty.")

---

### Exercise 35:
Write a Python function `double_each_element(lst)` that doubles each element in a list.

```python
# Exercise 35
```
def double_each_element(lst):
    """Double each element in the list."""
    return [element * 2 for element in lst]
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [float(x) for x in input_list]

result = double_each_element(numbers)
print(f"The list after doubling each element is: {result}")

---

### Exercise 36:
Write a Python function `mean(lst)` that returns the average of a list of numbers.

```python
# Exercise 36
```
def mean(lst):
    if not lst:  
        return 0  
    
    total = sum(lst)  
    average = total / len(lst)  
    return average
input_list = input("Enter numbers separated by spaces: ").split()
numbers = [float(x) for x in input_list]

result = mean(numbers)
print(f"The average of the list is: {result}")

---

### Exercise 37:
Write a Python function `remove_vowels(s)` that removes all vowels from a string.

```python
# Exercise 37
```
def remove_vowels(s):
    """Remove all vowels from the string."""
    vowels = "aeiouAEIOU"  
    result = "".join(char for char in s if char not in vowels)  
    return result
input_string = input("Enter a string: ")
result = remove_vowels(input_string)
print(f"The string after removing vowels is: '{result}'")

---

### Exercise 38:
Write a Python function `is_multiple_of(n, divisor)` that checks whether a number is a multiple of another number.

```python
# Exercise 38
```
def is_multiple_of(n, divisor):
    """Check if n is a multiple of divisor."""
    if divisor == 0:
        raise ValueError("Divisor cannot be zero.")  
    return n % divisor == 0
try:
    number = int(input("Enter a number (n): "))
    divisor = int(input("Enter a divisor: "))
    
    if is_multiple_of(number, divisor):
        print(f"{number} is a multiple of {divisor}.")
    else:
        print(f"{number} is n

---

### Exercise 39:
Write a Python function `concat_strings(lst)` that concatenates all strings in a list into a single string.

```python
# Exercise 39
```
def concat_strings(lst):
    """Concatenate all strings in the list into a single string."""
    return ''.join(lst)  
input_list = input("Enter strings separated by commas: ").split(',')
strings = [s.strip() for s in input_list]

result = concat_strings(strings)
print(f"The concatenated string is: '{result}'")

---

### Exercise 40:
Write a Python function `get_unique_elements(lst)` that returns a list of unique elements from a given list.

```python
# Exercise 40
```
def get_unique_elements(lst):
    """Return a list of unique elements from the given list."""
    unique_elements = list(set(lst))  
    return unique_elements
input_list = input("Enter elements separated by spaces: ").split()
result = get_unique_elements(input_list)
print(f"The unique elements are: {result}")

---
````
